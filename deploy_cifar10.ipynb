```python
import matplotlib.pyplot as plt
import numpy as np
import random
import keras
import glob
import cv2
```

```python
model = keras.models.load_model("cifar10env.model")
```

```python
def load_img(DIR):
    images=[]
    for image in glob.glob(DIR):
        img = cv2.imread(image)
        img = cv2.resize(img,(32,32))
        images.append(img)
    images = np.array(images)
    return images
Dir = "images\*"
images = load_img(Dir)
```

```python
predictions = model.predict(images)
```

```python
label=['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
i = random.randint(0,10)
plt.imshow(images[i])
plt.title(f'this is our {label[predictions[i].argmax()]}')
plt.show()
```
